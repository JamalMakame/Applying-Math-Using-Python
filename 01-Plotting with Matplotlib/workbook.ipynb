## Basic Plotting with Matplotlib


```python
import matplotlib.pyplot as plt
from typing import List, Tuple
import numpy as np

# A function that takes a list of x values and returns a list of y values


def input_function(x: List[int]) -> List[int]:
    return x*(x - 2) * np.exp(3-x)


# A list of x values
x = np.linspace(-0.5, 3.0, 100)

# A list of y values
y = input_function(x)

fig = plt.figure(figsize=(10, 6))

# Plot the function
plt.plot(x, y, x, x**2, x, 1-x)


# Add a title
plt.title("A Sample Function")
plt.xlabel("x = [-0.5, 3.0]")
plt.ylabel("y = x(x-2)e^(3-x)")
plt.grid(True)


# show the plot
plt.show()
```

## Changing the plotting style


```python
y1 = np.array([1, 2, 3, 4, 5])
y2 = np.array([1.2, 1.6, 3.1, 4.2, 4.8])
y3 = np.array([3.2, 1.1, 2.0, 4.9, 2.5])


plt.plot(y1, 'o', y2, 'x', y3, '*')
```

## Adding labels and Legnends to plots


```python
plt.plot(y1, 'o-', y2, 'x--', y3, '*-.')
plt.title("Plot of the data y1, y2 and y3")
plt.xlabel("x = [0, 4]")
plt.ylabel("y = [1, 5]")
plt.legend(["data y1", "data y2", "data y3"])
```

## Adding subplots


```python
def generate_newton_iters(x0: int, number: int) -> Tuple[List[float], List[float]]:
    iterates = [x0]
    errors = [abs(x0 - 1.)]
    for _ in range(number):
        x0 -= (x0*x0 - 1.)/(2*x0)
        iterates.append(x0)
        errors.append(abs(x0 - 1.))
    return iterates, errors
```

```python
iterates,  errors = generate_newton_iters(2.0, 5)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, tight_layout=True, figsize=(10, 4))
ax1.plot(iterates, 'x')
ax1.set_title("Iterates")
ax1.set_xlabel("$i$")
ax1.set_ylabel("$x_i$")

ax2.semilogy(errors, 'o')
ax2.set_title("Errors")
ax2.set_xlabel("$i$")
ax2.set_ylabel("Error")
```

## Surface and Contour plots

```python
X = np.linspace(-2, 2, 100)
Y = np.linspace(-1, 1, 100)


x,y = np.meshgrid(X, Y)

z= x**2 * y**3
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$z$")

ax.set_title("Graph of the function $f(x) = x^2y^3$")
```

```python
fig = plt.figure()  # Force a new figure
plt.contour(x, y, z)
plt.title("Contours of $f(x) = x^2y^3$")
plt.xlabel("$x$")
plt.ylabel("$y$")
```

## Customizing three-dimensional plots

```python
X = np.linspace(-2, 2)
Y = np.linspace(-2, 2)
x, y = np.meshgrid(X, Y)
t = x**2 + y**2  # small efficiency
z = np.cos(2*np.pi*t)*np.exp(-t)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

ax.set_title("Surface with colormap")
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$z$")
```

```python
fig = plt.figure(figsize=(16, 9))
plt.contour(x, y, z, cmap="viridis")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("Contour plot with colormap set")
```

```python
from matplotlib.colors import LightSource

light_source = LightSource(azdeg=0, altdeg=45)
cmap = plt.get_cmap("viridis")
vals = light_source.shade(z, cmap=cmap, vert_exag=0.1, blend_mode='soft')
surf = ax.plot_surface(x, y, z, rstride=1, cstride=1, facecolors=vals)
```
